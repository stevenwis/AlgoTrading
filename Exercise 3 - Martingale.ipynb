```python
# https://www.investopedia.com/articles/forex/06/martingale.asp
# https://en.wikipedia.org/wiki/St._Petersburg_paradox
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


C = 200
B = 20

T = 8
q = .7

win = np.zeros(T)
cash = np.zeros(T)
leverage = np.zeros(T)
bet = np.zeros(T)

leverage[0] = B/C
bet[0] = B
cash[0] = C

for i in range(1, T):
    
    bet[i] = B*np.power(2,i) 
    leverage[i] = bet[i]/cash[i-1]
    
    if np.random.rand() > q:
        cash[i] = cash[i-1] + bet[i]/2
        print("YOU WON THE MARTINGALE!")
        break
    else:
        cash[i] = cash[i-1] - bet[i]/2
        
    if cash[i] < 0:
        cash[i:] = cash[i]
        print("YOU WENT BANKRUPT!")
        break
        

plt.plot(range(0, T), bet)
plt.plot(range(0, T), cash)
plt.plot(range(0, T), np.zeros(T),'black')
plt.axvline(x=i,color= 'red')
plt.title('Cash - Bet')
plt.show()

plt.plot(range(0, T), leverage)
plt.axvline(x=i, color= 'red')
plt.title('Leverage')
plt.show()
```
